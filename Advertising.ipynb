# **Simple Linear Regression Project – Advertising Dataset**

This project performs a Simple Linear Regression analysis on the classic Advertising dataset.
The goal is to understand how advertising spending across three channels; TV, Radio, and Newspaper affects Sales.
The project builds three separate linear regression models;
1.   TV → Sales
2.   Radio → Sales
3.   Newspaper → Sales

Then it compares their performance based on:
*   Slopes (Sales return per advertising dollar → ROI)
*   R² Scores (predictive power)

This analysis is especially relevant for marketing analytics, business strategy, and advertising optimization.

# Importing the Library

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# **Section:1** Importing the Dataset (TV → Sales)


```python
dataset = pd.read_csv('Advertising_Data.csv')

X = dataset[['TV']]
y = dataset['Sales']
```

# Splitting the dataset into the Training set and Test set


```python
from sklearn.model_selection import train_test_split
X_train_tv, X_test_tv, y_train_tv, y_test_tv = train_test_split(X , y, test_size = 0.2, random_state = 1)
```

# Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor_tv = LinearRegression()
regressor_tv.fit(X_train_tv, y_train_tv)
```

# Predicting the Test set results

```python
y_pred_tv = regressor_tv.predict(X_test_tv)
```

# Visualizing the Training set results

```python
plt.scatter(X_train_tv, y_train_tv, color = 'red')
plt.plot(X_train_tv, regressor_tv.predict(X_train_tv), color = 'blue')
plt.title('Sales vs TV (Training set)')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```

# Visualizing the Test set results

```python
plt.scatter(X_test_tv, y_test_tv, color = 'red')
plt.plot(X_train_tv, regressor_tv.predict(X_train_tv), color = 'blue')
plt.title('Sales vs TV (Test set)')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```

# Checking if TV spending strongly increases Sales

```python
print("Coefficient (slope):", regressor_tv.coef_[0])
print("Intercept:", regressor_tv.intercept_)
```

# Evaluating the model using R² score

```python
from sklearn.metrics import r2_score

y_pred_tv = regressor_tv.predict(X_test_tv)

print("R² Score:", r2_score(y_test_tv, y_pred_tv))
```
